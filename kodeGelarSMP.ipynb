```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
%matplotlib inline
```

<h3> LOAD MAP </h3>

```python
#load map dataframe
map_df = gpd.read_file('map_indo/INDONESIA_PROP.shp')
```

```python
map_df['Propinsi'] = map_df['Propinsi'].replace('IRIAN JAYA BARAT', 'PAPUA BARAT')
```

```python
map_df['Propinsi'] = map_df['Propinsi'].replace('DAERAH ISTIMEWA YOGYAKARTA', 'DI YOGYAKARTA')
```

```python
map_df['Propinsi'] = map_df['Propinsi'].replace('NUSATENGGARA BARAT', 'NUSA TENGGARA BARAT')
```

```python
map_df['Propinsi'] = map_df['Propinsi'].replace('PROBANTEN', 'BANTEN')
```

```python
map_df['Propinsi'] = map_df['Propinsi'].replace(['IRIAN JAYA TIMUR', 'IRIAN JAYA TENGAH'], 'PAPUA')
```

```python
map_df['Propinsi'] = map_df['Propinsi'].replace('DI. ACEH', 'ACEH')
```

```python
map_df.rename(columns = {'Propinsi' : 'NAMA PROVINSI'}, inplace = True)
```

```python
plt.rcParams['figure.figsize'] = [60, 70]
map_df.plot()
```

```python
kalut = gpd.read_file('IDN_adm/IDN_adm1.shp')
df = kalut.loc[kalut['NAME_1'] == 'Kalimantan Utara']
df = df.drop(['ID_0','ISO','NAME_0','ID_1','TYPE_1','ENGTYPE_1','NL_NAME_1','VARNAME_1'], axis = 1)
```

```python
df.rename(columns = {'NAME_1' : 'NAMA PROVINSI'}, inplace = True)
```

```python
map_df = pd.concat([df, map_df])
```

```python
map_df['NAMA PROVINSI'] = map_df['NAMA PROVINSI'].replace('Kalimantan Utara', 'KALIMANTAN UTARA')
```

```python
kaltim = gpd.read_file('IDN_adm/IDN_adm1.shp')
kaltim = kaltim.loc[kaltim['NAME_1'] == 'Kalimantan Timur']
kaltim = kaltim.drop(['ID_0','ISO','NAME_0','ID_1','TYPE_1','ENGTYPE_1','NL_NAME_1','VARNAME_1'], axis = 1)
kaltim.rename(columns = {'NAME_1' : 'NAMA PROVINSI'}, inplace = True)
```

```python
map_df = pd.concat([kaltim, map_df])
```

```python
kepri = gpd.read_file('IDN_adm/IDN_adm1.shp')
kepri = kepri.loc[kepri['NAME_1'] == 'Kepulauan Riau']
kepri = kepri.drop(['ID_0','ISO','NAME_0','ID_1','TYPE_1','ENGTYPE_1','NL_NAME_1','VARNAME_1'], axis = 1)
kepri.rename(columns = {'NAME_1' : 'NAMA PROVINSI'}, inplace = True)
```

```python
map_df = pd.concat([kepri, map_df])
```

```python
map_df['NAMA PROVINSI'] = map_df['NAMA PROVINSI'].replace('Kepulauan Riau', 'KEPULAUAN RIAU')
```

```python
map_df = map_df.drop([20])
```

```python
map_df['NAMA PROVINSI'] = map_df['NAMA PROVINSI'].replace('Kalimantan Timur', 'KALIMANTAN TIMUR')
```

```python
df_un = pd.read_excel('smp.xlsx')
df_un.head()
```

```python
merged = map_df.merge(df_un, how='left', left_on="NAMA PROVINSI", right_on="NAMA PROVINSI")
```

```python
variable = 'RERATA NILAI'
vmin, vmax = 40,70
fig, ax = plt.subplots(1, figsize=(30, 10))
ax.axis('off')
sm = plt.cm.ScalarMappable(cmap='Greens', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm.set_array([])
fig.colorbar(sm)
merged.plot(column=variable, cmap='Greens', linewidth=0.8, ax=ax, edgecolor='0.8')

# add a title and annotation
ax.set_title('Visualisasi Nilai UN SMP', fontdict={'fontsize': '25', 'fontweight' : '3'})
ax.annotate('https://hasilun.puspendik.kemdikbud.go.id/', xy=(0.6, .05), xycoords='figure fraction', fontsize=12, color='#555555')
```

```python
# Add Labels
merged['coords'] = merged['geometry'].apply(lambda x: x.representative_point().coords[:])
merged['coords'] = [coords[0] for coords in merged['coords']]
for idx, row in merged.iterrows():
    plt.annotate(s=row['NAMA PROVINSI']
xy=row['coords'],horizontalalignment='center')
```
