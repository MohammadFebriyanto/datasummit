```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
%matplotlib inline
```

<h3> LOAD MAP </h3>

```python
#load map dataframe
map_df = gpd.read_file('map_indo/INDONESIA_PROP.shp')
map_df
```

```python
map_df['Propinsi'] = map_df['Propinsi'].replace('IRIAN JAYA BARAT', 'PAPUA BARAT')
```

```python
map_df['Propinsi'] = map_df['Propinsi'].replace('DAERAH ISTIMEWA YOGYAKARTA', 'DI YOGYAKARTA')
```

```python
map_df['Propinsi'] = map_df['Propinsi'].replace('NUSATENGGARA BARAT', 'NUSA TENGGARA BARAT')
```

```python
map_df['Propinsi'] = map_df['Propinsi'].replace('PROBANTEN', 'BANTEN')
```

```python
map_df['Propinsi'] = map_df['Propinsi'].replace(['IRIAN JAYA TIMUR', 'IRIAN JAYA TENGAH'], 'PAPUA')
```

```python
map_df['Propinsi'] = map_df['Propinsi'].replace('DI. ACEH', 'ACEH')
```

```python
map_df.rename(columns = {'Propinsi' : 'NAMA PROVINSI'}, inplace = True)
```

```python
map_df
```

```python
plt.rcParams['figure.figsize'] = [60, 70]
map_df.plot()
```

<h3> LOAD DATASET UN PROVINSI </h3>

```python
df1 = pd.read_excel('PROV_IPA.xlsx')
df2 = pd.read_excel('PROV_BAHASA.xlsx')
df3 = pd.read_excel('PROV_IPS.xlsx')
df4 = pd.read_excel('PROV_KEAGAMAAN.xlsx')
df5 = pd.read_excel('PROV_KATOLIK.xlsx')
df6 = pd.read_excel('PROV_PROTESTAN.xlsx')
```

```python
df_un = pd.concat([df1,df2,df3,df4,df5,df6])
```

```python
df_un = df_un.drop(df_un.index[0])
df_un = df_un[df_un['RERATA NILAI'] != '-']
df_un
```

```python
df_un.head(35)
```

```python
pd.to_numeric(df_un['RERATA NILAI']).describe()
```

```python
df_un = df_un.drop([35], axis = 0)
```

```python
df_un.head(34)
```

```python
merged = map_df.merge(df_un, how='left', left_on="NAMA PROVINSI", right_on="NAMA PROVINSI")
```

```python
merged.loc[merged['NAMA PROVINSI'] == 'DI YOGYAKARTA']
```

```python
variable = 'RERATA NILAI'
vmin, vmax = 0,100
fig, ax = plt.subplots(1, figsize=(30, 10))
ax.axis('off')
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm.set_array([])
fig.colorbar(sm)
merged.plot(column=variable, cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8')
```
